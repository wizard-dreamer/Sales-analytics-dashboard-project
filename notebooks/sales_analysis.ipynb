# Sales Data Analytics using SQL and Python

Beginner-friendly analysis pipeline: load data to SQLite, run SQL with pandas, and generate matplotlib charts.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline

# Connect to SQLite database
db_path = '../sales.db'  # relative to notebooks folder
conn = sqlite3.connect(db_path)
```

## 1. Load and validate data from SQL

```python
sales_df = pd.read_sql_query('SELECT * FROM sales', conn, parse_dates=['Order_Date'])
sales_df.head()
```

## 2. Analysis queries and charts

```python
# 2.1 Top selling products
q1 = '''
SELECT Product_Name, Category, SUM(Quantity) AS total_quantity, SUM(Sales) AS total_sales
FROM sales
GROUP BY Product_Name, Category
ORDER BY total_quantity DESC
LIMIT 10;
'''
top_products = pd.read_sql_query(q1, conn)
display(top_products)

plt.figure(figsize=(10,6))
plt.bar(top_products['Product_Name'], top_products['total_quantity'], color='tab:blue')
plt.xticks(rotation=45, ha='right')
plt.title('Top Selling Products by Quantity')
plt.ylabel('Quantity')
plt.tight_layout()
plt.savefig('../visuals/top_selling_products.png')
plt.show()
```

```python
# 2.2 Monthly revenue trend
q2 = '''
SELECT strftime('%Y-%m', Order_Date) AS month, SUM(Sales) AS monthly_revenue
FROM sales
GROUP BY month
ORDER BY month;
'''
monthly_trend = pd.read_sql_query(q2, conn)
display(monthly_trend)

plt.figure(figsize=(10,5))
plt.plot(monthly_trend['month'], monthly_trend['monthly_revenue'], marker='o', color='tab:green')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.savefig('../visuals/monthly_revenue_trend.png')
plt.show()
```

```python
# 2.3 Country-wise sales distribution
q3 = '''
SELECT Country, SUM(Sales) AS total_sales
FROM sales
GROUP BY Country
ORDER BY total_sales DESC;
'''
country_sales = pd.read_sql_query(q3, conn)
display(country_sales)

plt.figure(figsize=(8,8))
plt.pie(country_sales['total_sales'], labels=country_sales['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Country Sales Distribution')
plt.axis('equal')
plt.tight_layout()
plt.savefig('../visuals/country_sales_distribution.png')
plt.show()
```

```python
# 2.4 Profit by category
q4 = '''
SELECT Category, SUM(Profit) AS total_profit, SUM(Sales) AS total_sales
FROM sales
GROUP BY Category
ORDER BY total_profit DESC;
'''
profit_category = pd.read_sql_query(q4, conn)
display(profit_category)

plt.figure(figsize=(8,5))
plt.bar(profit_category['Category'], profit_category['total_profit'], color='tab:purple')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.savefig('../visuals/profit_by_category.png')
plt.show()
```

```python
# Close the DB connection
conn.close()
```
